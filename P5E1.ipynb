```python
import numpy as np
from random import random

def f_tinv():
  u = random()
  if u < 0.25:
    return 2 + 2*np.sqrt(u)
  else:
    return 6 - 2*np.sqrt(3*(1-u))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def f(x):
    return np.where(
        (2 <= x) & (x <= 3),
        (x - 2) / 2,
        np.where(
            (3 < x) & (x <= 6),
            (2 - x / 3) / 2,
            0
        )
    )

dominio = np.linspace(2, 6, 1000)
num_muestras = int(1e5)
muestras = np.empty(num_muestras)

for i in range(num_muestras):
    muestras[i] = f_tinv()

# Gráfico
sns.lineplot(x=dominio, y=f(dominio), color='red', label='f(x)')
sns.histplot(muestras, stat='density', kde=False, bins=50, label='Simulación')

plt.xlabel('x')
plt.ylabel('Densidad')
plt.title('Comparación entre f(x) y simulación inversa')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
