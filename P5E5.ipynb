```python
from random import random
import numpy as np
# a.i)

def maxX(generadores_de_muestras):
  return max([gen() for gen in generadores_de_muestras])

# a.ii)
def minX(generadores_de_muestras):
  return min([gen() for gen in generadores_de_muestras])

# b

def generar_muestra_exponencial(lambd):
  u = random()
  return -np.log(1-u)/lambd

gen_exp_l1 = lambda: generar_muestra_exponencial(1)
gen_exp_l2 = lambda: generar_muestra_exponencial(2)
gen_exp_l3 = lambda: generar_muestra_exponencial(3)

n = 10000
generadores = [gen_exp_l1, gen_exp_l2, gen_exp_l3]
muestras_maxX = [maxX(generadores) for _ in range(n)]
muestras_minX = [minX(generadores) for _ in range(n)]


import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(muestras_maxX, stat='density', color='green', label='max')
sns.histplot(muestras_minX, stat='density', color='red', label='min')

plt.xlim(0, 5) # grafico cortado
plt.legend()
plt.show()
```
