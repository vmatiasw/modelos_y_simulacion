```python
from random import seed, random

def muestreador_bootstrap(muestra):
    n = len(muestra)
    return [muestra[int(random() * n)] for _ in range(n)]

def varianza(xs, muestral=False):
    n = len(xs)
    media = sum(xs) / n
    dividendo = n - 1 if muestral else n
    return sum((x - media) ** 2 for x in xs) / dividendo

def estimar_varS2(muestra, B=1000):
    # Replicaciones bootstrap
    Scuad_replicas = [
        varianza(muestreador_bootstrap(muestra), muestral=True) for _ in range(B)
    ]

    # Varianza de las replicaciones (aprox de Var(S))
    return varianza(Scuad_replicas)

# Datos
MUESTRA_A = [1, 3]
MUESTRA_B = [5, 4, 9, 6, 21, 17, 11, 20, 7, 10, 21, 15, 13, 16, 8]

seed(1)
print("Muestra A")
varS2 = estimar_varS2(MUESTRA_A)
print(f"Var(S): {varS2:.4f}")

seed(1)
print("\nMuestra B")
varS2 = estimar_varS2(MUESTRA_B)
print(f"Var(S): {varS2:.4f}")
```
